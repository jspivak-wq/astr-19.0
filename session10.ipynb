Select your favorite distribution (other than uniform and normal) and pull 1000 random numbers from that distribution. Histogram the random numbers into 100 bins, and plot the histogram.
```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.chisquare(2,4,1000)
bins = 100
plt.hist(x, bins = bins)
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('hist.pdf')
```
